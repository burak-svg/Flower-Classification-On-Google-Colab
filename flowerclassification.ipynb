```python
# %% [markdown]
# # Çiçek Sınıflandırma Uygulaması
# Bu notebook, basit bir çiçek sınıflandırma modeli oluşturup test etmenizi sağlar.

# %% [markdown]
# ## 1. Gerekli Kütüphaneleri Yükleme
import numpy as np
import matplotlib.pyplot as plt
import os
import PIL
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import pathlib

# %% [markdown]
# ## 2. Veri Setini Yükleme ve Hazırlama
# Google Colab'de çalıştırıyorsanız, veri setini Google Drive'dan veya direkt indirerek kullanabilirsiniz.

# %%
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos', origin=dataset_url, untar=True)
data_dir = pathlib.Path(data_dir)

# %%
# Veri setindeki çiçek türleri
# Corrected: Navigate into the 'flower_photos' subdirectory to get class names
data_dir_flowers = data_dir / 'flower_photos'
class_names = sorted([item.name for item in data_dir_flowers.iterdir() if item.is_dir()])
print("Sınıflar:", class_names)

# %% [markdown]
# ## 3. Veri Setini Görselleştirme
# %%
# Örnek görüntüleri göster
plt.figure(figsize=(10, 10))
# Use len(class_names) to ensure we don't go out of bounds if there are fewer than 9 classes
num_images_to_show = min(9, len(class_names))
for i in range(num_images_to_show):
    plt.subplot(3, 3, i + 1)
    # Corrected: Access images from the correct class subdirectories using data_dir_flowers
    image_paths = list(data_dir_flowers.glob(f'{class_names[i]}/*'))
    if image_paths: # Check if there are images in the directory
        img = PIL.Image.open(image_paths[0])
        plt.imshow(img)
        plt.title(class_names[i])
        plt.axis("off")
    else:
        # Handle cases where a directory might be empty
        plt.text(0.5, 0.5, "No image found", horizontalalignment='center', verticalalignment='center')
        plt.title(class_names[i])
        plt.axis("off")
plt.show()

# %% [markdown]
# ## 4. Veri Artırma ve Ön İşleme
# %%
batch_size = 32
img_height = 180
img_width = 180

# Eğitim verisi için %80 kullanıyoruz
# Corrected: Use data_dir_flowers for image_dataset_from_directory
train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir_flowers,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

# Doğrulama verisi için %20 kullanıyoruz
# Corrected: Use data_dir_flowers for image_dataset_from_directory
val_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir_flowers,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

# %% [markdown]
# ## 5. Model Oluşturma
# %%
num_classes = len(class_names)

# Add data augmentation layers
data_augmentation = keras.Sequential(
  [
    layers.RandomFlip("horizontal",
                      input_shape=(img_height,
                                  img_width,
                                  3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
  ]
)


model = Sequential([
  data_augmentation, # Add data augmentation here
  layers.Rescaling(1./255), # Rescaling layer comes after augmentation
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model.summary()

# %% [markdown]
# ## 6. Model Eğitimi
# %%
epochs=25
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)

# %% [markdown]
# ## 7. Eğitim Sonuçlarını Görselleştirme
# %%
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()

# %% [markdown]
# ## 8. Modeli Test Etme
# %%
# Kendi resminizi yükleyip test edebilirsiniz
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  # Yüklenen resmi yükle ve ön işleme yap
  img_path = fn
  img = tf.keras.utils.load_img(img_path, target_size=(img_height, img_width))
  img_array = tf.keras.utils.img_to_array(img)
  img_array = tf.expand_dims(img_array, 0) # Batch boyutu ekle

  # Tahmin yap
  predictions = model.predict(img_array)
  score = tf.nn.softmax(predictions[0])

  # Sonucu göster
  plt.figure()
  plt.imshow(img)
  plt.axis('off')
  plt.title(f"{class_names[np.argmax(score)]} (%{100 * np.max(score):.2f} confidence)")
  plt.show()

# %% [markdown]
# ## 9. Modeli Kaydetme
# %%
# Modeli kaydetmek isterseniz:
model.save('flower_classifier.h5')
```
